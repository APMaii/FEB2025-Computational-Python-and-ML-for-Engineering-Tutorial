```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])

plt.plot(x,y)
plt.show()
```

```python
#------------------
'''

marker ---> che noee bashe pointet --> 'o' , '*' , 'x' , '+' 

marker moshakhasat ---> moishakhaste noghte---------
ms= yek adad midi --> size noghatye
mec ---> range birone noghte --> 'r' , 'b' ,....
mfc---> range dakhel



-----moshakahsdate khat------
ls -->linme style --> '-' , '--' , '.-' , ':'
linewidth--> size
colour , c --> 'r'



'''
plt.plot(x,y,marker='o',ms=20,mec='g',mfc='r',ls=':',linewidth=15,c='y')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
ax = plt.axes(projection='3d')

#-----dataye shoma hast ama man example mizrm-----

z=np.linspace(0,10,100)
x=np.sin(z)
y=np.cos(z)

#---------------

ax.scatter3D(x,y,z,'gray',s=1)
ax.set_xlabel('sin')
ax.set_ylabel('cos')
ax.set_zlabel('value')

```
